```python
import numpy as np
import pandas
import pandas as pd
import seaborn
from matplotlib import pyplot as plt
```

```python
data = pandas.read_csv('AmesHousing.csv', index_col="Order")
print(data)
```

```python
print(data.keys())
print(data[['PID', 'SalePrice']].head())
print(data['Overall Qual'].head())
```

```python
training_features = [
    'Overall Qual', 'Gr Liv Area', 'Total Bsmt SF', 'Garage Cars', 'Garage Area', 
    'Year Built', 'Year Remod/Add', 'Full Bath', 'Neighborhood', 'MS Zoning', 
    'Lot Frontage', 'Lot Area', 'Sale Type', 'Mo Sold', 'Yr Sold', 'Condition 1'
]

selected_data = data[training_features]
print(selected_data.head())
```

```python
target_feature = data['SalePrice']
```

```python
# Function to map categorical features to int
def generate_ordinal_encoding(column):
    values = column.unique()
    mapping = {i: values[i] for i in range(len(values))}
    return mapping

def generate_ordinal_encoding_for_all_columns(dataset):
    encoding = {}
    for column in dataset.columns:
        if dataset[column].dtype == object:
            encoding[column] = generate_ordinal_encoding(dataset[column])
    return encoding

def apply_ordinal_encoding_to_all_columns(dataset, columns_dict):
    for column in columns_dict:
        dataset[column] = dataset[column].map({v: k for k, v in columns_dict[column].items()})

def apply_onehot_encoding(column):
    encoding = {}
    for feature in column.unique():
        encoding[feature] = column.apply(lambda x: x == feature)
    return pandas.DataFrame(encoding, index=column.index)
```

```python
# Convert 'Neighborhood' column to one-hot encoding
one_hot_neighborhood = pd.get_dummies(selected_data['Neighborhood'], prefix='Neighborhood')

# Drop the original 'Neighborhood' column and concatenate the one-hot encoded columns
selected_data = selected_data.drop('Neighborhood', axis=1)
selected_data = pd.concat([selected_data, one_hot_neighborhood], axis=1)

print(selected_data.head())
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler

# Convert categorical columns to one-hot encoding
selected_data_encoded = pd.get_dummies(selected_data, columns=['MS Zoning', 'Sale Type', 'Condition 1'])

# Identify numerical features
numerical_features = ['Lot Area', 'Gr Liv Area', 'Total Bsmt SF', 'Garage Area']

# Apply StandardScaler
scaler = StandardScaler()
selected_data_encoded[numerical_features] = scaler.fit_transform(selected_data_encoded[numerical_features])

target_feature = np.log1p(target_feature)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(selected_data_encoded, target_feature, test_size=0.2, random_state=42)


# Initialize the XGBoost regressor
xgb_model = XGBRegressor(objective='reg:squarederror', n_estimators=500, learning_rate=0.05, max_depth=5, subsample=0.8, colsample_bytree=0.8, random_state=42)

# Fit the model on the training data
xgb_model.fit(X_train, y_train)

# Predict
y_test_pred = xgb_model.predict(X_test)
y_train_pred = xgb_model.predict(X_train)

# Calculate the mean squared error
mse = mean_squared_error(y_test, y_test_pred)
print(f"Mean Squared Error: {mse}")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_test_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.title('True Values vs Predictions')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Compute residuals
residuals = y_test - y_test_pred

# Plot residual distribution
plt.figure(figsize=(8,5))
sns.histplot(residuals, bins=30, kde=True)
plt.axvline(x=0, color="red", linestyle="dashed")
plt.xlabel("Residuals")
plt.title("Residual Distribution")
plt.show()
```

```python
# Scatter plot of residuals
plt.figure(figsize=(8,5))
plt.scatter(y_test_pred, residuals, alpha=0.5)
plt.axhline(y=0, color="red", linestyle="dashed")
plt.xlabel("Predicted Prices")
plt.ylabel("Residuals (Errors)")
plt.title("XGBoost Residuals vs Predicted Prices")
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))

mae_train = mean_absolute_error(y_train, y_train_pred)
mae_test = mean_absolute_error(y_test, y_test_pred)

r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)

print(" XGBoost Regression Model Summary ")
print(f" Train RMSE: {rmse_train:.2f} | Test RMSE: {rmse_test:.2f}")
print(f" Train MAE: {mae_train:.2f} | Test MAE: {mae_test:.2f}")
print(f" Train R: {r2_train:.4f} | Test R: {r2_test:.4f}")
```

```python
feature_importance = pd.DataFrame({"Feature": selected_data_encoded.columns, "Coefficient": xgb_model.feature_importances_})
feature_importance = feature_importance[feature_importance["Coefficient"] != 0]  # Remove zero coefficients
feature_importance = feature_importance.sort_values(by="Coefficient", ascending=False)

# Plot top 10 features
plt.figure(figsize=(10,6))
sns.barplot(x=feature_importance["Coefficient"][:10], y=feature_importance["Feature"][:10], palette="Blues_r")
plt.xlabel("Feature Weight")
plt.ylabel("Feature")
plt.title("Top Important Features")
plt.show()
```

```python
# Compute baseline RMSE (Predicting the average house price)
y_baseline_pred = np.full_like(y_test, y_train.mean())
baseline_rmse = np.sqrt(mean_squared_error(y_test, y_baseline_pred))

print(f"Baseline RMSE (Mean Price Prediction): {baseline_rmse:.2f}")
print(f"XGBoost RMSE: {rmse_test:.2f}")

improvement = 100 * (baseline_rmse - rmse_test) / baseline_rmse
print(f"XGBoost Model Improves RMSE by: {improvement:.2f}%")
```
